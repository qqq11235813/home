```python
import random
import gym
import numpy as np
from collections import deque
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
class DQNAgent:
    def __init__(self, 
                 env, 
                 gamma = 0.95,
                epsilon = 1.0,
                epsilon_min = 0.01,
                epsilon_decay = 0.995,
                learning_rate = 0.001,
                batch_size = 32
                ):
        self.state_size = env.observation_space.shape[0]
        self.action_size = env.action_space.n
        self.memory = deque(maxlen=2000)
        self.gamma = gamma    # discount rate
        self.epsilon = epsilon  # exploration rate
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.end_when_done = True
        self.model = self._build_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse',
                      optimizer=Adam(lr=self.learning_rate))
        return model

    def data2info(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        act_values = self.model.predict(state)
        return np.argmax(act_values[0])  # returns action

    def update(self):
        minibatch = random.sample(self.memory, self.batch_size)
        for state, action, reward, next_state, done in minibatch:
            target = reward
            if not done:
                target = (reward + self.gamma *
                          np.amax(self.model.predict(next_state)[0]))
            target_f = self.model.predict(state)
            target_f[0][action] = target
            self.model.fit(state, target_f, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        self.model.load_weights(name)

    def save(self, name):
        self.model.save_weights(name)
        
    def obv2state(self, observation):
        return np.reshape(observation, [1, self.state_size])
    
    def update_judge(self):
        if len(self.memory) > self.batch_size:
            return True
        else:
            return False
```

```python
def train(env_name = "CartPole-v1",
          episode_num = 1000,
          max_timestep = 200,
          punish_r = -10,
          gamma = 0.95,    # discount rate
            epsilon = 1.0,  # exploration rate
            epsilon_min = 0.01,
            epsilon_decay = 0.995,
            learning_rate = 0.001,
            batch_size = 32
         ):
    performance = []
    
    env = gym.make(env_name)
    
    agent = DQNAgent(env,
                    gamma = gamma,    # discount rate
                    epsilon = epsilon,  # exploration rate
                    epsilon_min = epsilon_min,
                    epsilon_decay = epsilon_decay,
                    learning_rate = learning_rate,
                    batch_size = batch_size)
    
    for e in range(episode_num):
        observation = env.reset()
        state = agent.obv2state(observation)
        
        for t in range(max_timestep):
            action = agent.act(state)##############################
            next_observation, reward, done, _ = env.step(action)
            reward = punish_r if done else reward
            
            next_state = agent.obv2state(next_observation)
            agent.data2info(state, action, reward, next_state, done)
            
            state = next_state
            
            if (done and agent.end_when_done):
                print("episode {}: end at {}t".format(e, t + 1))
                performance.append(t+1)
                break
            
        if(agent.update_judge()):
            agent.update()
    return performance 
```

```python
from sklearn.model_selection import ParameterGrid
import time
def text_editor(para):
    text = ''
    for name in para:
        text += name 
        text += ':'
        text += str(para[name])
        text += '; '
    return text

def pf_machine():
    parameters = {'env_name' : ["CartPole-v1"],
                  'episode_num': [500],
                  'max_timestep' :[200],
                  'punish_r' : [-100],
                  'gamma' : [0.95],    # discount rate
                    'epsilon' : [1.0],  # exploration rate
                    'epsilon_min' : [0.01],
                    'epsilon_decay' : [0.995],
                    'learning_rate' : [0.001],
                    'batch_size' : [32]
                }
    
    seeds = [1,20,4000]
    
    for para in ParameterGrid(parameters):
        pfs = []
        for seed in seeds:
            random.seed(seed)
            np.random.seed(seed)
            text = text_editor(para)
            print('current seed', seed)
            print("parameter: ", text)
            
            cur_time = prev_time = time.time()
            
            performance = train(
                                env_name = para['env_name'],
                                  episode_num = para['episode_num'],
                                  max_timestep = para['max_timestep'],
                                  punish_r = para['punish_r'],
                                  gamma = para['gamma'],    # discount rate
                                    epsilon = para['epsilon'],  # exploration rate
                                    epsilon_min = para['epsilon_min'],
                                    epsilon_decay = para['epsilon_decay'],
                                    learning_rate = para['learning_rate'],
                                    batch_size = para['batch_size']
                                 )
            pfs.append(performance)
            
            print('one total training cost {}t'.format(time.time()- prev_time))
        np.save('data/'+text, pfs)
    
```

```python
pf_machine()
```
