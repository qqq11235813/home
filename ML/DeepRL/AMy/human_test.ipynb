```python
import gym
import matplotlib.pyplot as plt
import keyboard
import time
```

```python
def get_key_action(previous_action):
    if keyboard.is_pressed('q'):
        return 0
    elif keyboard.is_pressed('w'):
        return 1
    else:
        return previous_action
    
def test():
    if keyboard.is_pressed('p'):
        return True
    else:
        return False
```

```python
def human_gym(env_name = 'CartPole-v0', episode = 100, game_length = 1000, speed = 1):
    env = gym.make('CartPole-v0')
    score = []
    flag = True
    for i_episode in range(episode):
        if(keyboard.is_pressed('m')):
            break
        time.sleep(1)
        #flag = test()
        if(flag):
            episode += 1
            observation = env.reset()
            previous_action = 0
            for t in range(game_length):
                time.sleep(0.1)
                env.render()
                #action = env.action_space.sample()
                previous_action = action = get_key_action(previous_action)
                observation, reward, done, info = env.step(action)
                if done:
                    print("Episode finished after {} timesteps".format(t+1))
                    score.append(t+1)
                    break
    plt.plot(score)
    plt.show()
    return score
```
