```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def delta(n, g):
    #return 4*np.pi*n*h*g/lamda0
    return 0.5*np.pi*g

def r_org(k, n_all):
        return (n_all[k]-n_all[k-1]+0.0)/(n_all[k]+ n_all[k-1])
    
def transform(n_all, g, k=0):
    
    if(k == (len(n_all)- 2)):
        return (r_org(k, n_all)+ r_org(k+1, n_all)* np.exp(1j* delta(n_all[k], g)))/(1.0+ r_org(k, n_all)* r_org(k+1, n_all)* np.exp(1j* delta(n_all[k], g)))
    
    return (r_org(k, n_all)+ transform(n_all, g, k+1)* np.exp(1j* delta(n_all[k], g)))/(1.0+ r_org(k, n_all)* transform(n_all, g, k+1)* np.exp(1j* delta(n_all[k], g)))
def miu(n):
    alpha = 1.0
    return alpha* n

def mat_sp(n, g):
    delta_t = delta(n, g)
    #print "delta", delta_t
    mat = [[np.cos(delta_t), 1j* np.sin(delta_t)/miu(n)],[1j* np.sin(delta_t)* miu(n),np.cos(delta_t)]]
    mat = np.mat(mat)
    #print mat
    return mat

def transform_mat(n_all, g):
    k_t = len(n_all)- 2
    mat_all = np.mat([[1.0,0.0],[0.0,1.0]])
    for k in range(k_t):
        mat_all = np.dot(mat_sp(n_all[k+1], g),mat_all) 
        #print mat_all
    r = (mat_all[0,0]* miu(n_all[0])+ mat_all[0,1]* miu(n_all[0])* miu(n_all[-1])- mat_all[1,0]- mat_all[1,1]* miu(n_all[-1]))/(mat_all[0,0]* miu(n_all[0])+ mat_all[0,1]* miu(n_all[0])* miu(n_all[-1])+ mat_all[1,0]+ mat_all[1,1]* miu(n_all[-1]))
    return pow(abs(r), 2)
    #return r
```

# matrix method

```python
n_all = [1.0,2.38,1.38,2.38,1.38,2.38,1.38,2.38,1.38,2.38,1.38,2.38,1.5]
gg = np.arange(0.0, 2.0, 0.01)
R_g = []
import time
start = time.time()
for g_ele in gg:
    R_g.append(transform_mat(n_all, g_ele))
end= time.time()
plt.plot(gg, R_g)
plt.show()
print "matrix using " + str(end - start) + " seconds"
```
